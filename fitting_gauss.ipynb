## Adjust and calculate Gaussian fitting

```python
y0 = y[:128]
y1 = y[128:256]
y2 = y[256:411]
y3 = y[411:]

gauss_ans = []

x0 = np.array([i for i in range(len(y0))])
#param_init0 = [1400,40,10]
#popt1,pcov0 = curve_fit(gauss_func, x0, y0, param_init0, maxfev=100000)
g0 = lf.line_fitted(x0, y0)
gauss_ans += list(g0)

x1 = np.array([i for i in range(len(y1))])
#param_init1 = [1400,40,10]
#popt1,pcov1 = curve_fit(gauss_func, x1, y1, param_init1, maxfev=100000)
g1 = lf.line_fitted(x1, y1)
gauss_ans += list(g1)

x2 = np.array([i for i in range(len(y2))])
#param_init2 = [2000,50,100]
#popt2,pcov2 = curve_fit(gauss_func, x2, y2, param_init2, maxfev=1000000)
g2 = lf.line_fitted(x2, y2)
gauss_ans += list(g2)

x3 = np.array([i for i in range(len(y3))])
#param_init3 = [500,15,10]
#popt3, pcov3 = curve_fit(gauss_func, x3, y3, param_init3, maxfev=1000000)
g3 = lf.line_fitted(x3, y3)
gauss_ans += list(g3)

gauss_ans_japan = gauss_ans
```

## Plot Gaussian fitting curv

## confficient of Determination（決定係数)

```python
mean_y = np.mean(y)
sum_y = sum_y_hat = 0

for i in range(len(y)):
    sum_y_hat += (y[i]-gauss_ans_japan[i])*(y[i]-gauss_ans_japan[i])

for i in range(len(y)):
    sum_y += (y[i]-mean_y)*(y[i]-mean_y)

R2 = 1-(sum_y_hat/sum_y)
print(R2)
```

## 区間ごとの決定係数

```python
error_y = []
error_y_hat = []

for i in range(len(y)):
    error_y.append((y[i]-mean_y)*(y[i]-mean_y))

for i in range(len(y)):
    error_y_hat.append((y[i]-gauss_ans_japan[i])*(y[i]-gauss_ans_japan[i]))

R2_1 = 1-(sum(error_y_hat[:159])/sum(error_y[:159]))
R2_2 = 1-(sum(error_y_hat[159:257])/sum(error_y[159:257]))
R2_3 = 1-(sum(error_y_hat[257:353])/sum(error_y[257:353]))
R2_4 = 1-(sum(error_y_hat[353:])/sum(error_y[353:]))
print(R2_1)
print(R2_2)
print(R2_3)
print(R2_4)
```
